##  Problématique

**Peut-on prédire automatiquement si un Pokémon est légendaire à partir de ses caractéristiques ?**

Les Pokémon légendaires représentent environ 12% du dataset, ce qui en fait un problème de classification déséquilibrée. L'objectif est de comparer plusieurs modèles prédictifs pour identifier celui qui détecte le mieux cette classe minoritaire.

## Préprocessing (`Prepro.py`)

Tous les notebooks de modélisation partagent une fonction commune `prepare_data()` :

- Nettoyage des noms de colonnes
- Suppression des colonnes non pertinentes (`name`, `abilities`, `bmi`...)
- Séparation features / cible
- `StandardScaler` sur les variables numériques
- `OneHotEncoder` sur les variables catégorielles
- Split train/test stratifié (80/20)

## Modèles comparés

### Tableau comparatif des 3 modèles

| | Random Forest | Gradient Boosting | SVC-RBF |
|---|---|---|---|
| **Accuracy** | 99% | 99% | 97% |
| **F1 Légendaires** | 94% | <span style="color: #e91e1e;">96%</span> | 86% |
| **Sensibilité Légendaires** | 92% | <span style="color: #e91e1e;">96%</span>  | 84% |
| **Précision Légendaires** | 96% | 96% | 88% |
| **Légendaires trouvés** | 23/25 | <span style="color: #e91e1e;">24/25</span>  | 21/25 |

### Classement

1. **Gradient Boosting** : meilleur sur tous les critères, détecte 24/25 légendaires

2. **Random Forest** : très proche, légèrement moins bon sur le recall (23/25)

3. **SVC-RBF** : en retrait, rate 4 légendaires sur 25 et F1 de seulement 86%


### Conclusion générale 

> Les trois modèles obtiennent de bonnes performances globales, mais se distinguent nettement sur la détection des légendaires.
>
> Le Gradient Boosting s'impose comme le meilleur modèle avec un F1 de 96% et 24 légendaires sur 25 correctement identifiés. Il tire parti de sa stratégie séquentielle pour corriger les cas difficiles. Le Random Forest est très proche (F1 94%) et offre en plus une analyse plus riche des variables importantes. Le SVC-RBF, bien que correct en accuracy globale (97%), peine davantage sur les légendaires (F1 86%, 4 ratés) —> il est moins adapté à ce déséquilibre de classes.
>
> Dans tous les modèles, le **catch_rate** ressort comme la variable la plus discriminante, confirmant que les légendaires sont avant tout définis par leur rareté dans le jeu plutôt que par leurs stats de combat.
>
