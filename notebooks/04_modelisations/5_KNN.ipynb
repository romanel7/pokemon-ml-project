# KNN

### ## Pourquoi faire KNN malgré ses limites dans notre dataset ?

KNN ne sera pas notre meilleur modèle.
Mais il apporte quelque chose que Random Forest n'apporte pas : il raisonne
par **similarité**.

Concrètement, au lieu d'apprendre des règles ("si BST > 580 alors..."),
KNN se demande juste : *"est-ce que ce Pokémon ressemble aux Légendaires
que j'ai déjà vus ?"*

Le problème c'est qu'avec peu de Légendaires et beaucoup de colonnes,
cette notion de "ressemblance" devient floue.

Pour y remédier, on va utiliser uniquement les features que le Random
Forest a identifiées comme importantes. Moins de colonnes = des distances
plus significatives = un KNN plus efficace.
